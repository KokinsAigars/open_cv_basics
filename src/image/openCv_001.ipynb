```python
new_img = cv2.resize(img, (1000,400))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img_2 = cv2.resize(img, (0,0), img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img_2)
```

```python
new_img_3 = cv2.flip(new_img_2, 0)
```

```python
plt.imshow(new_img_3)
```

```python
new_img_4 = cv2.flip(new_img_2, 1)
```

```python
plt.imshow(new_img_4)
```

```python
# flip bouth axes
new_img_5 = cv2.flip(new_img_2, -1)
plt.imshow(new_img_5)
```

```python
type(new_img_5)
```

```python
# save image 
cv2.imwrite('../output/00-puppy_fliped.jpg', new_img_5)
```
