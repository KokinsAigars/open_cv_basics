# drawing on top of images

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16)
```

```python
type(blank_img)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(384,10),pt2=(500,150),color=(0,255,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(200,200),pt2=(300,300),color=(50,45,10),thickness=5)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(100,100),radius=50,color=(250,0,0),thickness=8)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(400,400),radius=50,color=(250,0,0),thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(0,0),pt2=(450,450),color=(255,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
#text on image
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='text on image',org=(10,500),fontFace=font,fontScale=1,color=(255,0,255),
            thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
# cutom shape on image
```

```python
blank_img_2 = np.zeros(shape=(512,512,3),dtype=np.int32)
```

```python
plt.imshow(blank_img_2)
```

```python
vertices = np.array([ [100,300], [200,200], [400,300], [200,400] ],dtype=np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
# need to reshape my vertices so they have 3 planes rgb
pts = vertices.reshape((-1,1,2))
```

```python
# the same dimaension, just one more plane 
pts.shape
```

```python
pts
```

```python
cv2.polylines(blank_img_2,[pts],isClosed=True,color=(250,0,0),thickness=5)
plt.imshow(blank_img_2)
```
