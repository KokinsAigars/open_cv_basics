# BLENDING IMAGES OF THE SAME SIZES

```python
# distortion will creep in
img_rgb1 = cv2.resize(img_rgb1, (1200,1200))
img_rgb2 = cv2.resize(img_rgb2, (1200,1200))
```

```python
blended = cv2.addWeighted(src1=img_rgb1,alpha=0.5,src2=img_rgb2,beta=0.5,gamma=0)
```

```python
plt.imshow(blended)
```

```python
## OVERLAY SMALL IMAGE ON TOP OF LARGER IMAGE
```

```python
img1 = cv2.imread('/app/output/00-puppy.jpg')
img_rgb1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('/app/output/watermark_no_copy.png')
img_rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img_rgb2 = cv2.resize(img_rgb2,(600,600))
```

```python
large_img = img_rgb1
small_img = img_rgb2
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end,x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```
