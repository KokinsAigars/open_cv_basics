Image-Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/app/output/00-puppy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img = cv2.imread('/app/output/00-puppy.jpg',0)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img,cmap='gray')
```

```python
img
```

```python
### Binary
ret,thresh1 = cv2.threshold(img,170,255,cv2.THRESH_BINARY)
plt.imshow(thresh1,cmap='gray')
```

```python
ret
```

```python
### Binary Inverse
ret,thresh2 = cv2.threshold(img,17,255,cv2.THRESH_BINARY_INV)
plt.imshow(thresh2,cmap='gray')
```

```python
###  Threshold Truncation
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
plt.imshow(thresh3,cmap='gray')
```

```python
### Threshold to Zero
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
plt.imshow(thresh4,cmap='gray')
```

```python
### Threshold to Zero (Inverse)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh5,cmap='gray')
```

# Real World Applications

## Adaptive Thresholding



### Sudoku Image

```python
img = cv2.imread("/app/output/crossword.jpg",0)
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
    ax.axis('off')
```

```python
show_pic(img[80:400, 200:550])
```

```python
## Simple Binary
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
show_pic(th1[80:400, 200:550])
```

```python
### Adaptive Threshold
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8) # Play around with these last 2 numbers
show_pic(th2[80:400, 200:550])
```

```python
blended = cv2.addWeighted(src1=th1,alpha=0.7,src2=th2,beta=0.3,gamma=0)
show_pic(blended[80:400, 200:550])
```
